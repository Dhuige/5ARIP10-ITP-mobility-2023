```python
import pandas as pd
import numpy as np
import os

data_path = "D:/荷兰/TUe/课程/Team Project/Dataset/highd-dataset-v1.0/data"
n_files = 60
map_ids = [1, 2, 3, 4, 5, 6]
output_path = "output"

if not os.path.exists(output_path):
    os.makedirs(output_path)

for map_id in map_ids:
    print(f"processing map {map_id}...\r", end="")

for i in range(n_files):
    # Load tracks and tracksMeta files
    tracks_path = os.path.join(data_path, f"{i+1:02d}_tracks.csv")
    tracks_meta_path = os.path.join(data_path, f"{i+1:02d}_tracksMeta.csv")
    tracks = pd.read_csv(tracks_path)
    tracks_meta = pd.read_csv(tracks_meta_path)

    # Merge tracks and tracksMeta files to get driving direction
    merged = pd.merge(tracks, tracks_meta[["id", "class", "drivingDirection"]], on="id")

    # Split tracks based on class and driving direction
    cars_dir1 = merged.loc[(merged["class"] == "Car") & (merged["drivingDirection"] == 1)]
    cars_dir2 = merged.loc[(merged["class"] == "Car") & (merged["drivingDirection"] == 2)]
    trucks_dir1 = merged.loc[(merged["class"] == "Truck") & (merged["drivingDirection"] == 1)]
    trucks_dir2 = merged.loc[(merged["class"] == "Truck") & (merged["drivingDirection"] == 2)]

    # Calculate speed for each frame
    data_dict = {"cars_dir1": cars_dir1, "cars_dir2": cars_dir2, "trucks_dir1": trucks_dir1, "trucks_dir2": trucks_dir2}
    for key, data in data_dict.items():
        data["speed"] = np.sqrt(data["xVelocity"]**2 + data["yVelocity"]**2) * 3600 / 1000
        data = data[["id", "frame", "speed"]]
        data.to_csv(os.path.join(output_path, f"{i+1:02d}_{key}_speed.csv"), index=False)

print("Done!")
```
