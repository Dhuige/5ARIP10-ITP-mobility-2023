```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
# Read files to dictionary
data_path = ".\..\Data\highd-dataset-v1.0\data"
n_files = 60
data_dict = {}
for i in range(n_files):
    print(f"reading file {i+1}/{n_files}...\r", end="")
    file_name = f"{i+1:02d}_tracks.csv"
    file_path = os.path.join(data_path, file_name)
    data_dict[f"data_{i+1}"] = pd.read_csv(file_path)
```

```python
# Plot TTCs
frameRate = 25
frameTime = 1/frameRate

plt.figure(figsize=(10,6))
for i in range(n_files):
    frames = data_dict[f"data_{i+1}"]["ttc"].shape[0]
    times = np.linspace(1,frameTime*frames,frames)
    plt.plot(times,data_dict[f"data_{i+1}"]["ttc"], label=f"Vehicle {i+1}");
plt.xlabel("time (s)");
plt.ylabel("TTC (s)");
```
